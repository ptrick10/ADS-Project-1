```python
# preprocess external dataset
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import folium
```

```python
'''
Data from: https://www.census.gov/data/datasets/time-series/demo/popest/2010s-counties-total.html
'''


populationData = pd.read_csv("../Proj1data/nycCountyData.csv", index_col=False)
```

```python
populationData = populationData.iloc[3:66]
```

```python
populationData.columns = ['Geographical Area', 'Census', 'Estimates Base', '2010', '2011', '2012', '2013', '2014', '2015', '2016', '2017', '2018', '2019',]
```

```python
populationData.set_index(populationData["Geographical Area"])
```

```python
print(populationData.head())
```

```python
State_population = populationData.iloc[0, 3:]
print(State_population)
```

```python
'''
The five counties for NYC are Kings, Bronx, Richmond, Queens, and New York counties, 
(as seen on this website: https://www.ny.gov/counties)
'''
# find 5 counties
bronx = populationData.iloc[3, 3:]
print(bronx)
kings = populationData.iloc[24, 3:]
print(kings)
ny = populationData.iloc[31, 3:]
print(ny)
queens = populationData.iloc[41, 3:]
print(queens)
richmond = populationData.iloc[43, 3:]
print(richmond)

City_population = bronx + kings + ny + queens + richmond
print(City_population)
```
